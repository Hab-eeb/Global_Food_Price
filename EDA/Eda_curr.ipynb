```python
#importing of necessary libraries and packages
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Importing and reading the data
#The data imported has been cleaned by removing the first 5 rows because they were non essential info and reduced the actual number of columns
curr_data = pd.read_csv("data/currency.csv", sep= '\t')
curr_data.head()
```

```python
curr_data.columns
```

```python
#Isolating the Nigerian data 
curr_data[curr_data["Country Name"] == "Nigeria"]
```

```python
nig_curr = curr_data[curr_data["Country Name"] == "Nigeria"]
```

```python
nig_curr
```

```python
nig_curr.isnull().sum()
```

```python
#Dropping the first three columns 
num_nig = nig_curr.drop(["Country Name","Country Code", "Indicator Name","Indicator Code"], axis=1)

num_nig.head()
```

```python
num_nig_T = num_nig.T
```

```python
num_nig_T.rename(columns = {172:"1#-1$"},inplace =True)
num_nig_T.head()
```

```python
plt.figure(figsize=(50,10))
plt.plot(num_nig_T.index, num_nig_T["1#-1$"] , color='green', linestyle='dashed', linewidth = 1,
         marker='o', markerfacecolor='blue', markersize=7)
plt.xlabel('Years',fontsize=25)
plt.ylabel('Naira(#)',fontsize=25)
  

plt.title('Official Naira to Dollar Exchange rate 1960-2020 ',fontsize=30)
plt.yticks(fontsize=20)
plt.xticks(fontsize=15)
  

plt.show()
```
